```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('data/qc_performance.xlsx')
```

```python
df.Worst = df.Worst * -1
```

```python
df = df.reset_index().rename(columns={'index':'Dataset'}).sort_values(['Best','Worst'], ascending=False)
```

```python
sns.set(style="whitegrid", context="talk")
fig, axs = plt.subplots(1,2, figsize=(5,3), dpi=300)
wax = axs[0]
bax = axs[1]
sns.barplot(x="Worst", y="Dataset", data=df, ax=wax, color="#d6604d")
sns.barplot(x="Best", y="Dataset", data=df, ax=bax, color="#4393c3")
wax.set_yticklabels([])
wax.set_ylabel("")
wax.set_xlabel("# Metrics Ranked Worst")

bax.set_xlabel("# Metrics Ranked Best")
bax.set_ylabel("")
for tick in bax.yaxis.get_major_ticks():
    tick.label1.set_horizontalalignment('center')
    tick.label1.set_x(-0.15)
fig.tight_layout(pad=0)
fig.savefig("metrics_best_worst.png")
```
